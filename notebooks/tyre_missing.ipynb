```python
df = pd.read_csv('tyre_sales_clean.csv')
```

```python
df.head(31)
```

```python
for col in ['price', 'units_sold']:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())
```

```python
text_cols = ['region', 'tyre_type']

for col in text_cols:
    if col in df.columns:
        df[col] = df[col].fillna('Unknown')
```

```python
df['total_sales'] = df['total_sales'].fillna(df['total_sales'].median())
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.dtypes
```

```python
df['date'] = pd.to_datetime(df['date'])
df['tyre_size'] = pd.to_numeric(df['tyre_size'], errors='coerce')
df['units_sold'] = pd.to_numeric(df['units_sold'], errors='coerce')
df['unit_price'] = pd.to_numeric(df['unit_price'], errors='coerce')
df['total_sales'] = pd.to_numeric(df['total_sales'], errors='coerce')
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')

```

```python
df.isnull().sum()
```

```python
df[['units_sold','unit_price','total_sales','revenue']].describe()
```

```python
df['units_sold'] = df['units_sold'].clip(lower=df['units_sold'].quantile(0.01),
                                         upper=df['units_sold'].quantile(0.99))

df['unit_price'] = df['unit_price'].clip(lower=df['unit_price'].quantile(0.01),
                                         upper=df['unit_price'].quantile(0.99))

df['total_sales'] = df['total_sales'].clip(lower=df['total_sales'].quantile(0.01),
                                           upper=df['total_sales'].quantile(0.99))

df['revenue'] = df['revenue'].clip(lower=df['revenue'].quantile(0.01),
                                   upper=df['revenue'].quantile(0.99))
```
