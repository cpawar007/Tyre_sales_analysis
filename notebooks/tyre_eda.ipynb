```python
df = pd.read_csv('tyre_sales_cleaned_final.csv')
```

```python
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.to_period('M')
monthly_sales = df.groupby('month')['revenue'].sum()

monthly_sales.plot(kind='line', figsize=(10,5))
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.title('Monthly Tyre Sales Trend')
plt.show()
```

```python
region_sales = df.groupby('region')['revenue'].sum()

region_sales.plot(kind='bar', figsize=(8,5))
plt.xlabel('Region')
plt.ylabel('Revenue')
plt.title('Region-wise Revenue')
plt.show()
```

```python
tyre_type_sales = df.groupby('tyre_type')['units_sold'].sum()

tyre_type_sales.plot(kind='bar', figsize=(8,4))
plt.xlabel('Tyre Type')
plt.ylabel('Units Sold')
plt.title('Tyre Type Performance')
plt.show()
```

```python
top_sizes = df.groupby('tyre_size')['units_sold'].sum().sort_values(ascending=False).head(10)

top_sizes.plot(kind='bar', figsize=(8,4))
plt.xlabel('Tyre Size')
plt.ylabel('Units Sold')
plt.title('Top 10 Best Selling Tyre Sizes')
plt.show()
```
